```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from feature_engineer import get_engineered_df




DATA_PATH = "~/Lucas_Systems_Capstone_Project/data/processed/oe_detailed.parquet"
df, features, cat_cols = get_engineered_df(DATA_PATH, warehouse="OE", max_time=300, work_code="30")

```

```python
# y_ij = mu + a_i (task/location) + b_j (worker) + e_ij
model = smf.mixedlm(
    "Time_Delta_sec ~ 1",          # fixed effect: just the intercept (grand mean)
    data=df,
    groups=df["UserID"],           # one random effect via groups
    exog_re=np.ones(len(df))       # scalar random intercept per worker
)
result = model.fit()
print(result.summary())
```
