```python
rows = []

for id in dfsmall["UserID"].unique():
    for date in dfsmall["Date"].unique():
        row_data = {"UserID": id, 
                    "Date": date, 
                    "Number of AssignmentOpen": (dfsmall[dfsmall['UserID'] == id][dfsmall["Date"] == date][dfsmall["ActivityCode"] == "AssignmentOpen"]).shape[0],
                    "Number of PickPut": (dfsmall[dfsmall['UserID'] == id][dfsmall["Date"] == date][dfsmall["ActivityCode"] == "PickPut"]).shape[0],
                    "Number of AssignmentID": (dfsmall[dfsmall['UserID'] == id][dfsmall["Date"] == date]["AssignmentID"]).nunique()}

        rows.append(row_data)

ao_rows = pd.DataFrame(rows)
```
