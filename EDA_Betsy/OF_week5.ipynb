```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
of_df = pd.read_parquet("/Users/betsyfrdmn/Desktop/Lucas Code/Lucas_Systems_Capstone_Project/data/processed/of_joined.parquet")
```

```python
of_df.Level.value_counts().sort_index(ascending=True)
```

```python
# Create a pivot table of the average Pick_Time for each Aisle and Level
# We use 'Level' if you have a column for it; otherwise, use 'Bay'
pivot_df = of_df.pivot_table(
    index='Level', 
    columns='Aisle', 
    values='Time_Delta_sec', 
    aggfunc='mean'
).sort_index(ascending=False)
  
plt.figure(figsize=(30, 8))
sns.heatmap(pivot_df, annot=True, cmap='YlOrRd', fmt='.0f')
plt.title('Heatmap of Average Pick Time (Hours) by Aisle and Level')
plt.show()
```

```python
# Use 'std' to find the Standard Deviation
pivot_std = of_df.pivot_table(
    index='Level', 
    columns='Aisle', 
    values='Time_Delta_sec', 
    aggfunc='std'
).sort_index(ascending=False)

plt.figure(figsize=(30, 8))
sns.heatmap(pivot_std, annot=True, cmap='Purples', fmt='.0f')
plt.title('Heatmap of Pick Time Variability (Standard Deviation in Seconds)')
plt.show()
```
